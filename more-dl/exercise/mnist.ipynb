```python
import torchvision.datasets as dts
from torchvision.transforms import ToTensor
import torch
import matplotlib.pyplot as plt
import random

train_set = dts.MNIST(
    root = 'data',
    train = True,                         
    transform = ToTensor(), 
    download = True,            
)
test_set = dts.MNIST(
    root = 'data', 
    train = False, 
    transform = ToTensor()
)
print("DONE")

train_loader = torch.utils.data.DataLoader(train_set, batch_size=64, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_set, batch_size=64, shuffle=False)
data_iterator = iter(train_loader)
images, labels = next(data_iterator)


image = random.choice(images)
plt.imshow(image.numpy().squeeze(), cmap='gray');
```

```python
import torch.nn as nn

class MODEL(nn.Module):
    def __init__(self):
     super().__init__()
     self.layers = nn.Sequential(
            nn.Linear(784,16),
            nn.ReLU(),
            nn.Linear(16,16),
            nn.ReLU(),
            nn.Linear(16,10),   
         )
    def forward(self, x: torch.Tensor):
        x = x.view(-1, 784)
        return self.layers(x)
    
```

```python
import torch.optim as optim
model = MODEL() 
lossFunction = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr = 0.1)
```

```python
def train(train_loader, test_loader, optimizer, lossFunction, model, epochs):
    epochs = 10
    epochsCount, trainLoss, testLoss = [], [], []
    for epoch in range(epochs):
        model.train()
        for X,y in train_loader:
            output = model(X)
            print(0)
            loss =lossFunction(output,y)
            print(1)
            
            loss.backward()
            
            optimizer.step()
            optimizer.zero_grad()
    print("xxxxxxx")        
    model.eval()
            
    for i,j in test_loader:
        with torch.inference_mode():
            y_preds = model(X)
            test_loss = lossFunction(y_preds, labels)
        
    if epoch % 10 ==0:
        epochsCount.append(epoch)
        trainLoss.append(loss)
        testLoss.append(test_loss)
        print(f"Epoch {epoch}: train loss {loss}, test loss {test_loss}")
    return epochsCount, trainLoss, testLoss

    
train(train_loader, test_loader, optimizer, lossFunction, model,10 )               
            
            
```
